```python
data.drop(columns=['index'],inplace=True)
```

```python
data['Base Pay'].value_counts()      #we need to delete $
```

```python
data['Base Pay'] = data['Base Pay'].str.replace('$', '',regex=False)     #replace the  $ and set the '' 
```

```python
data['Base Pay'].dtypes
```

```python
data['Base Pay'] = data['Base Pay'].astype('float')       #change the datatype to float
```

```python
data[[ 'Name','Base Pay','Position']].sort_values(by=['Base Pay'] , ascending=False) [:30]     #sort the data by Base pay
```

```python
sns.barplot(data=data, y='University or Office', x='Base Pay')
```

from this conclusion of barplot in office of the chancellor has the highest count

```python
plot.scatter(x='Base Pay',y='University or Office',data=data)
plot.xlabel('Base Pay')
plot.ylabel('University or office')
plot.title('Variabe University or office')
plot.show()
```

In this graph it show that the west chester has the highest base pay.

```python
sns.catplot(x = "University or Office", y = "Base Pay", palette = "Set2", data = data,height=6,aspect=3)
```

In this graph office of the Chancellor has the highest base pay

```python
import plotly.graph_objects as go
import plotly.express as exp
```

```python
from re import X
fig = exp.bar(data, y='Name', x='Base Pay')
fig.update_layout(title='Name on base pay',paper_bgcolor='rgb(233,233,233)', yaxis_title='Name', xaxis_title='Base Pay', height=400, width=1000,showlegend=True)
fig.show()
```

```python
legend_groups=['Position','University or Office']
```

```python
legend_groups
```

```python
fig = go.Figure(layout=go.Layout(height=600, width=800, title=go.layout.Title(text="Position University & Pay")))
fig.add_trace(go.Scatter(x=data['Base Pay'], y=data['Position'], hovertext=['Base Pay', 'Position'], name=legend_groups[0]))
fig.add_trace(go.Scatter(x=data['Base Pay'], y=data['University or Office'], hovertext=['Base Pay', 'University or Office'], name=legend_groups[1]))
fig.update_layout(xaxis_title='Base Pay', yaxis_title='Position name')
fig.show()
```

In this graph it shows that instuctor position has highest base pay.
